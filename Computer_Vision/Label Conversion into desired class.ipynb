```python
import os
def update_yolo_labels(labels_dir, old_class, new_class):
    old_class_str = str(old_class)
    new_class_str = str(new_class)
    for filename in os.listdir(labels_dir):
        if filename.endswith('.txt'):
            file_path = os.path.join(labels_dir, filename)
            with open(file_path, 'r') as file:
                lines = file.readlines()
            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if parts and parts[0] == old_class_str:
                    parts[0] = new_class_str
                updated_lines.append(' '.join(parts))
            with open(file_path, 'w') as file:
                file.write('\n'.join(updated_lines) + '\n')
            print(f"Updated file: {filename}")
labels_directory = r"D:\FYP Documents\BALL BAT PICH DATASET\Ball-250MB\valid\labels"
old_class_index = 0
new_class_index = 1
update_yolo_labels(labels_directory, old_class_index, new_class_index)
```
